```python
# Cell 3: Import all required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from glob import glob

from sklearn.model_selection import train_test_split
from sklearn import metrics

import cv2
import os

import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

import warnings
warnings.filterwarnings('ignore')
```

```python
# Cell 4: Display sample images
# Define the path to the image sets and the class names
base_path = 'lung_colon_image_set/lung_image_sets'
classes = ['lung_n', 'lung_aca', 'lung_scc'] # Normal, Adenocarcinoma, Squamous Cell Carcinoma

for cat in classes:
    image_dir = f'{base_path}/{cat}'
    images = os.listdir(image_dir)

    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(f'Images for {cat} category . . .', fontsize=16)

    for i in range(3):
        k = np.random.randint(0, len(images))
        img = np.array(Image.open(f'{base_path}/{cat}/{images[k]}'))
        ax[i].imshow(img)
        ax[i].axis('off')
    plt.show()
```

```python
# Cell 5: Process images and create train/validation sets
# Define constants
IMG_SIZE = 256
SPLIT = 0.2
EPOCHS = 10
BATCH_SIZE = 64

# Prepare the data
X = []
Y = []

for i, cat in enumerate(classes):
    # Use glob to find all jpeg images in the directory
    image_files = glob(f'{base_path}/{cat}/*.jpeg')
    for image_path in image_files:
        try:
            img = cv2.imread(image_path)
            # Resize the image
            img_resized = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            X.append(img_resized)
            Y.append(i)
        except Exception as e:
            print(f"Error reading {image_path}: {e}")

# Convert lists to NumPy arrays
X = np.asarray(X)
# One-hot encode the labels
one_hot_encoded_Y = pd.get_dummies(Y).values

# Split the data
X_train, X_val, Y_train, Y_val = train_test_split(X, one_hot_encoded_Y,
                                                  test_size=SPLIT,
                                                  random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Validation data shape: {X_val.shape}")
```

```python
# Cell 6: Define the CNN model architecture
model = keras.models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(5, 5), activation='relu',
                  input_shape=(IMG_SIZE, IMG_SIZE, 3), padding='same'),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
    layers.Dense(3, activation='softmax') # 3 classes, so 3 output neurons
])

# Print the model summary
model.summary()
```

```python
# Cell 7: Compile the model and set up callbacks

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Define callbacks
es = EarlyStopping(patience=3, monitor='val_accuracy', restore_best_weights=True)
lr = ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.5, verbose=1)

class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('val_accuracy') > 0.95: # High threshold for stopping
            print('\nValidation accuracy has reached over 95%, stopping training.')
            self.model.stop_training = True
```

```python
# Cell 8: Train the model
history = model.fit(X_train, Y_train,
                    validation_data=(X_val, Y_val),
                    batch_size=BATCH_SIZE,
                    epochs=EPOCHS,
                    verbose=1,
                    callbacks=[es, lr, myCallback()])
```

```python
# Cell 9: Plot training and validation accuracy
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['accuracy', 'val_accuracy']].plot(title="Accuracy vs. Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
# Cell 10: Evaluate the model and print classification report
# Generate predictions on the validation set
Y_pred_probs = model.predict(X_val)

# Convert predicted probabilities and one-hot encoded true labels to class indices
Y_pred = np.argmax(Y_pred_probs, axis=1)
Y_val_indices = np.argmax(Y_val, axis=1)

# Print the classification report
print("Classification Report:\n")
print(metrics.classification_report(Y_val_indices, Y_pred, target_names=classes))
```
